## Steps in modeling:
1. ***Creating a model:*** Define the input and the output layer, as well as the hidden layers of a deep learning model
2. ***Compiling a model:*** Defining the Loss Function, the Optimizer and the Evaluation Matrix.
3. ***Fitting the model:*** Letting the model try to find pattern between X and y.

```python
import keras
```

SHIFT + CTRL + SPACE ----->  Bring out the Doc String

```python
# Model Building


# Set the random seed for "Reproducibility"
tf.random.set_seed(69)


# 1. Creating the model using sequential api
model = keras.Sequential([
                          keras.layers.Dense(1) # We can pass/create a architecture of a Neural Network; as elements of list or using model.add()
                        ])

# 2. Compile the model
model.compile(loss = keras.losses.mae, # MAE is a loss function; MAE --> Mean Absolute Error
              optimizer = keras.optimizers.SGD(), # SGD --> Stochastic Gradient Descent
              metrics = ["mae"] # Evaluation matrix
              )

# 3. Fit the Model
model.fit(X, y, epochs=10)
```

```python
y_pred = model.predict(
    [17.0] # For 17 the output value should be 27;
)
y_pred # Since the value of y_pred is very off..... as the mae is very high.
```

```python
y_pred + 12 # As we can see that the even-if we add the mae with the y_pred; still y_pred is very off; that suggests that our model is bad.
```

### Another way of creating the model: `model.add()`

```python
model2 = keras.Sequential()
model2.add(keras.layers.Dense(1))
model2.compile(loss = keras.losses.mae, optimizer = keras.optimizers.SGD(), metrics= ["mae"])
model2.fit(X, y, epochs = 10)
```

### Improving the NN model:
We create a model in  three steps.

1. **Creating a Model:** We can add more layer, So it becomes **more deep**, We can **incerease the number of Hidden Unit** ; We can even **add activation function**.

2. **Compiling a model:** We can change optimization function, we can change **Learning rate** of the optimization function.

3. **Fitting model:** We can add **more epochs**, or we canm feed **more Data** to the model.

#### Increase the model's epochs

```python
# Create a model
model = keras.Sequential([keras.layers.Dense(1)])

# Compile a Model
model.compile(loss= keras.losses.mae, optimizer=keras.optimizers.SGD(), metrics= ["mae"])

# Fitting a model larger epoches (Accuracy/evaluation matrix will overshoot)
model.fit(X, y, epochs= 100)
```

### üí† With a ploting function

```python
# Lets a create a ploting function
def plot_prediction(train_data = X_train, train_labels = y_train, test_data = X_test, test_labels = y_test, predictions = y_pred):
    """ train_data = X_train, train_labels = y_train, test_data = X_test, test_labels = y_test, predictions = y_pred;
    > > > This functions plots the training data and compares it with the ground truth labels. """
    
    # Size of the plot
    plt.figure(figsize = (16, 9))

    # Ploting the training data @(Blue color)
    plt.scatter(train_data, train_labels, color = "blue", label = "Training Data -->")

    # Plotting the testing data @(red Color)
    plt.scatter(test_data, test_labels, color = "green", label = "Testing Data -->")

    #ploting the model Predictions
    plt.scatter(test_data, predictions, color = "red", label = "Predictions" )

    # Showing the legends
    plt.legend();
```
