### Python methode inside a class

However, you must be wondering why we passed self as the parameter to the method description().

self is a self-referencing pointer. It is necessary to pass self as a parameter if the method is inside a class. This is because when an object calls a method, Python automatically passes the object as the first argument to the method.  
If we dont use self:  
    6   
      7 rec = Rectangle()  
----> 8 rec.description()  

`TypeError: description() takes 0 positional arguments but 1 was given`

```python
class Rectangle:
    length = 20
    width = 10
    def description(self):
        print("This is a rectangle.")

rec = Rectangle()
rec.description()
```

Lets pass `self` and another `parameter` 

```python
class Rectangle:
    length = 20
    width = 10
    def description(self):
        print("This is a rectangle.")
    
    def area(self, arm):
        return 10*arm

rec = Rectangle()
print(rec.description())
print(rec.area(3))
#print(rec.area(6,3))
```
