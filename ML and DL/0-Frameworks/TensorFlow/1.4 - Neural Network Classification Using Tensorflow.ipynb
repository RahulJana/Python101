## Basic Intuition of a Classification Model/Problem:
Let's say we are doing a Image classification,
* ***Binary Classifiaction:*** Image classification() cat vs Dog

* ***Multi Class Classification:*** (More than two class) Cat Vs Dog Vs Bird Image, Flower Classication is example of Multi-Class Classification, Each Image contains only one class.  
> Email Classifier is an example of Multi Class Classification. {Office, SHopping, Spam}
* ***Multi Label CLassification:*** Image Tagging(An Image contains both Cat and Dog, and we have to add two labels or tags to the image then it is a multi label Classification.)
> Movie review tagging could be an example of Multi Label Classification

* > **Some common parameters for Input Image:**  
1. **Batch_size:** 32(According to Processors architecture batch size is determined, 32 is satndard and Default batch size of the Tensorflow.) or None(None Refers to Batch size is not set, It could be a arbirary number). *Ammount of Images the system Processor(CPU or GPU) can look at.*

2. **Width:** Width of an Image.
3. **Height:** Height of an Image.
4. **Color Channel:** Generally 3.

* > ***Output Shape:***  
* > > let we have 3-classes at multi Class: 1. Cat; 2. Dog; 3. Bird.  Then the output shape is [3] ----> Example [0.04, 0.9, 0.06]; The Image is Dog.

## Creating a model for creating a visualization of decision boundary:
* ***Recipe:*** 
1. https://cs231n.github.io/neural-networks-case-study/
2. https://github.com/madewithml/guides
> * Take a trained model, features(X) and labels(y)
> * Create meshgrid of different X values
> * Make predictions accross a mesh grid [Link of Mesh Grid](https://www.geeksforgeeks.org/numpy-meshgrid-function/)
> * Plot the predictions as well as a line b/w zones

```python
# Creating function for ploting a decision boundary
def plot_decission_boundary(model, X, y):
    """
    Plots the decision boundary created by model predictions
    this fn is borrowed from:---------->
    https://cs231n.github.io/neural-networks-case-study/ and 
    https://github.com/madewithml/guides
    """

    # Define the axis boundaries of the plot and create a meshgrid
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                         np.linspace(y_min, y_max, 100))
    
    # Create X value (We will be making predictions on these)
    x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D-arrays together

    # Make Predictions
    y_pred = model.predict(x_in)

    # Check for multi-class
    if len(y_pred[0])> 1:
        print("Multiclass Classification.")
        # Since this a multi-class classification
        y_pred = np.round(y_pred).reshape(xx.shape)
    else:
        print("Binary Classification,")
        y_pred = np.round(y_pred).reshape(xx.shape)
    
    # Plot Decision boundary
    plt.figure(figsize=(7,7))
    plt.contourf(xx, yy, y_pred, cmap = plt.cm.RdYlBu, alpha = 0.7)
    plt.scatter(X[:, 0], X[:, 1], c = y, s = 40, cmap = plt.cm.RdYlBu)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
```
