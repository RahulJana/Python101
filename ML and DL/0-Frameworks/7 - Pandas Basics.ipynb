# Code Debugging

```
def factorial(x):
    if (x == 0):
        return 1
    return x*factorial(x-1)
```

```
factorial(4)
```

```
import random
import numpy as np
```

```
# This code snippet contains the error
def code_to_debug():
    for i in range(10):
        x = random.random()
        factorial(x)
```

```
code_to_debug()
# By calling this fn() we will create an error which is called max-recurssion-deapth error
```

### In this part we will solve the error

**We will use the python debugger for debugging**

```
def copy_of_code_to_debug():
    import pdb; pdb.set_trace()

    for i in range(10):
        x = random.random()
        factorial(x)
```

```
# Debugged factorial code:
def factorial_debugged(x):
    if(not isinstance(x, int)):
        # This if statement checks <if x is int or not>
        print("This method only supports int datatype.")
        return -1
    if(x== 0):
        return 1;
    return x*factorial(x-1)
```
