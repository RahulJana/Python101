```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
R = np.random.random([100, 1])
```

```python
R
```

```python
plt.plot(R)
plt.show()
```

```python
np.mean(R)
```

```python
np.std(R)
```

$z = \frac{x - u}{s}$

```python
z = (R - (np.mean(R)))/(np.std(R))
z
```

```python
plt.plot(z)
```

```python
# Instansiating the StandardScaler()
std_scaler = StandardScaler()
```

```python
std_scaler.fit(R)
```

```python
std_scaler.mean_
```

```python
# Transforming the real Data R
RT = std_scaler.transform(R)
```

```python
RT == z
```

```python
np.mean(RT)
```

```python
plt.plot(RT)
```

- In Standardaization the mean will become 0 and standard deviation will become 1.
- In Normalization, everything will be scaled down between 0-1 

### Standardaization for training-and testing datadet

As we have already notice that `StandardScaler`; has two functions `.fit()` and `.transform()`; So we `fit` our model at the training data and `transform` the training data & we will `only transform` the testing data. So the code Snippet looks like this.  
`X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state = seed)`  
`scaler = StandardScaler()`  
`X_scaled_train = scaler.fit_transform(X_train)`  
`X_scalde_test = scaler.transform(X_test)`  

* Now we can't transform the `Y` as in a neural network the value of y will reamin between `0 - 1` so we `can't use standardization` ; as in standardization the mean becomes 0 and std. deviaton will be 1.
