## 2-D Plotting of Sigmoid

***Sigmoid Function:***  
$S_{w,b}(x) = \frac{1}{1+e^{-(wx+b)}}$

```python
def sigmoid(x, w, b):
    return 1/(1+np.exp(-(w*x + b)))
```

```python
sigmoid(1, 0.5, 0)
```

```python
# Plotting of the Sigmoid Function

# Defining Parameters for the function
w = 0.9
b = 0
X = np.linspace(-10, 10, 100) # 100 points between -10 and 10

#Applying the Sigmoid function
y = sigmoid(X, w, b)

plt.plot(y)
plt.show()
```

```python
# function for output of sigmoid and plotting.
def sigmoid_plot(x, w, b):
    y = 1/(1+(np.exp(-(w*x + b))))
    plt.plot(y)
    plt.show()
```

```python
sigmoid_plot(X, w, b)
```

```python
'''
In the function sigmoid every parameter is scaler;
but when we pass a matrix/array to the function then python will do type casting;
and create vector/matrix; as it will iterate through every single value.
'''
# Scaler Componenet
temp_S = sigmoid(1, 1, 10)
print(type(temp_S))
print(type(y))
```

As we can see for `temp_S` type is `float`; but for `y` the datatype is `np.array`.

```python
w_unknown = 0.5
b_unknown = 0.25

X = np.random.random(100)*20 - 10
Y = sigmoid(X, w_unknown, b_unknown)
```

```python
"""
W = np.linspace(-1, 1, 100)
B = np.linspace(-1, 1, 100)
"""
# As we can see that that the -ve values are not working well;
# we will shirnk our search space and exclude our -ve values

W = np.linspace(0, 1, 100)
B = np.linspace(0, 1, 100)


WW, BB = np.meshgrid(W, B)
```

```python
print(WW.shape, BB.shape)
```

```python
def calculate_loss(X, Y, w_est, b_est):
    loss = 0
    for x, y in zip(X, Y):
        loss += (y - sigmoid(x, w_est, b_est))**2
    return loss
```

```python
total_loss = np.zeros(WW.shape)
total_loss.shape
```

```python
for i in range(WW.shape[0]):
    for j in range(WW.shape[1]):
        total_loss[i, j] = calculate_loss(X, Y, WW[i, j], BB[i, j])
```

### 3D plot of loss function: Using plot surface.

```python
def loss_plot_3D(X, Y, w, b, number_of_points = 500, elevation = 60, azimuthal_angles = 30):
    total_loss = np.zeros(WW.shape)
    
    for i in range(w.shape[0]):
        for j in range(w.shape[1]):
            total_loss[i, j] = calculate_loss(X, Y, w[i, j], b[i, j])
    
    # Finding the minimum loss
    ij =  np.argmin(total_loss)
    print("minimum loss:",ij)
    
    # Specific indecies
    i = int(np.floor(ij/total_loss.shape[1]))
    j = int(ij - i * total_loss.shape[1])
    print("i:", i, "j:", j)
    print("w[i, j]:", w[i, j], ";","b[i, j]:", b[i, j])
    
    # Plotting part
    plt.figure(figsize=(12, 8))
    ax = plt.axes(projection = "3d")
    ax.plot_surface(w, b, total_loss, cmap = "cool")
    ax.set_xlabel("WW")
    ax.set_ylabel("BB")
    ax.set_zlabel("Total Loss")
    ax.view_init(elevation, azimuthal_angles)
```

## Class for Sigmoid Neuron

```python
class SigmoidNeuron():
    def __init__(self):
        self.w = None
        self.b = None
    
    def perceptron(self, x):
        return np.dot(x, self.w.T) + self.b
    
    def sigmoid(self, x):
        return 1.0/(1.0 + np.exp(-(x)))
    
    def grad_w(self, x, y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y) * y_pred * (1 - y_pred) * x

    def grad_b(self, x, y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y) * y_pred * (1 - y_pred)
    
    def fit(self, X, Y, epochs = 1, learning_rate = 1, initialize = True):
        # Initialize w, b; if initialize == True then the loop will run
        if initialize:
            self.w = np.random.randn(1, X.shape[1])
            self.b = 0

        for i in range(epochs):
            db = 0
            dw = 0
            for x, y in zip(X, Y):
                dw += self.grad_w(x, y)
                db += self.grad_b(x, y)
            
            self.w -= learning_rate * dw
            self.b -= learning_rate * db
```

### Fit for toy data

```python
X = np.asarray([[2.5, 2.5], [4, -1], [1, -4], [-3, 1.25], [-2, -4], [1, 5]])
Y = [1,1,1,0,0,0]
```

```python
sn = SigmoidNeuron()
```

```python
sn.fit(X, Y, 1, 0.25, True)
```

```python
print(sn.w, sn.b)
```

```python
def plot_sn(X, Y, sn, ax):
    X1 = np.linspace(-10, 10, 100)
    X2 = np.linspace(-10, 10, 100)
    XX1, XX2 = np.meshgrid(X1, X2)
    YY = np.zeros(XX1.shape)
    for i in range(X2.size):
        for j in range(X1.size):
            val = np.asarray([ X1[j], X2[i] ])
            YY[i, j] = sn.sigmoid(sn.perceptron(val))
    ax.contourf(XX1, XX2, YY, cmap = "cool", alpha = 0.72)
    ax.scatter(X[:, 0], X[:, 1], c = Y, cmap = "viridis")
    ax.plot()
```

```python
plt.figure(figsize= (30, 10))

# instead of assigning weights we will run it through a for loop
sn.fit(X, Y, 1, 0.75, True)
N = 10
for i in range(N):
    print(sn.w, sn.b)
    ax = plt.subplot(2, 5, i+1)# We are using i+1 as subplot will plot from index i = 1; instead of i = 0
    plot_sn(X, Y, sn, ax)
    sn.fit(X, Y, learning_rate = 0.75, initialize = False) # We don't want the w and b to be re-initialized multiple time.
```
