# Reading image using cv2

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('car.jpg',cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('car.jpg',cv2.IMREAD_GRAYSCALE)
RGBimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.plot([0,400],[0,400],'g', linewidth = 5)
plt.imshow(RGBimg)
cv2.imwrite('roses.png',img)
```

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
cap = cv2.VideoCapture('video.mp4')
while(True):
  ret, frame = cap.read()
  cv2_imshow(frame)
  if cv2.waitKey(1) & 0xFF == ord('r'):
    break
cap.release()
cv2.destroyAllWindows()    
```

```
import numpy as np
import cv2
cap = cv2.VideoCapture('video.mp4')
while(True):
  ret,frame = cap.read()
  graying = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
  cv2_imshow(graying)
  if cv2.waitKey(1) & 0xFF == ord('r'):
    break 
cap.release()
cv2.destroyAllWindows()    
```
