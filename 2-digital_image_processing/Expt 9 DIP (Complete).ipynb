### Liner Transformation in SD
Image negative transformation S=(L-1)-r S=(256-1)-r S=255-r

```
# With GreyScale Image
import cv2
from google.colab.patches import cv2_imshow
img_1= cv2.imread('Grey.jpg',0)
img_2= 255 - img_1
cv2_imshow(img_1)
cv2_imshow(img_2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```
# With Colour Image
img_1= cv2.imread('car.jpg')
img_2= 255 - img_1
cv2_imshow(img_1)
cv2_imshow(img_2)
```

### Power law transformation
S=cr^gamma


```
img_1= cv2.imread('Grey.jpg',0)
gamma=1
img_2= np.power(img_1,gamma)
gamma=1.5
img_3= np.power(img_1,gamma)
cv2_imshow(img_1)
cv2_imshow(img_2)
cv2_imshow(img_3)
```
