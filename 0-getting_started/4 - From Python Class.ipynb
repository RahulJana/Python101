# Classes in python

```
# Classes in python
class MobilePhone:
    '''This the method of creation of the doc string of the class MobilePhone:
    We can put anything in this doc string 

    To access this documentation: 
            class_name.__doc__
    
    '''
    def __init__(self, name, is_android = False, screen_size = 5.5):
        '''
        __init__ -> is a built-in function
        When we pass self in the <<def __init__(self, Other_Attributes)>>
        it will instantiate the object(Which will be ignored in the end).
        But the next attribute that is passed(Here name) will be considered as
        an attribute.

        name -> String;
        is_android -> Boolean Value. (By default false, is user does not specifies then it is false)
        screen_size -> Float Value. 5.5" bydefault
        '''
        self.name = name
        self.is_android = is_android
        self.screen_size = screen_size
        self.rating = -1
        self.popularity = 0


    def has_rating(self):
        '''
        This is not a built-in function, Still we have to pass self as this is mandetory.
        has_rating returns boolean value.
        '''
        return self.rating > -1
    

    def has_popularity(self, popularity = 0):
        '''
        has_popularity returns integer value
        '''
        return (self.popularity * 1.1)
        
```
