```
tensor_multiplication = x*y
print(tensor_multiplication)
```

# NumPy vs PyTorch

```
# torch -> numpy
print(x)
print(x.dtype)

x_np = x.numpy()
print(x_np)
print(x_np.dtype)
```

```
# numpy -> torch
a = np.random.randn(6)
print(a)
print(type(a))

a_pt = torch.from_numpy(a)
print(a_pt)
print(a_pt.dtype)
```

```
temp = np.random.randint(2, 10, (3,2))
print(temp, type(temp))
```

## Speed Difference b/w numpy and pytorch

```
%%time
for i in range(10000):
    a = np.random.randn(100, 100)
    b = np.random.randn(100, 100)
    c = a+b
```

```
%%time
for i in range(10000):
    a = torch.randn(100, 100)
    b = torch.randn(100, 100)
    c = a+b
```

Well ......Torch is fast AF❤❤


---



Lets extend further

```
%%time
for i in range(10000):
    a = np.random.randn(100, 100)
    b = np.random.randn(100, 100)
    c = a*b
```

```
%%time
for i in range(10000):
    a = torch.randn(100, 100)
    b = torch.randn(100, 100)
    c = a*b
```

Matrix Multiplication

```
%%time
for i in range(10000):
    a = np.random.randn(100, 100)
    b = np.random.randn(100, 100)
    c = np.matmul(a, b)
```

```
%%time
for i in range(10000):
    a = torch.randn(100, 100)
    b = torch.randn(100, 100)
    c = torch.matmul(a, b)
```

```
a = torch.ones(3, 2, device= cuda0)
b = torch.ones(3, 2, device= cuda0)
c = a+b

print(a)
print(b)
print(c)
```

```
%%time
for i in range(10000):
    a = np.random.randn(100, 100)
    b = np.random.randn(100, 100)
    c = a*b
```

```
# torch on CPU
%%time
for i in range(10000):
    a = torch.randn(100, 100)
    b = torch.randn(100, 100)
    c = a*b
```

```
# torch on GPU
%%time
for i in range(10000):
    a = torch.randn([100, 100], device = cuda0)
    b = torch.randn([100, 100], device = cuda0)
    c = a*b
```
